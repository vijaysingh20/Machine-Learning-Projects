# 7. Significance Test:

### 1. Significance Test 1:

#### H0: concavity error doesn't effect Cancer Type   
#### H1: concavity error effects Cancer Type

```python
x = df['concavity error'].to_numpy().reshape(-1, 1)
y = df['Cancer_type'].to_numpy().reshape(-1, )
reg = LinearRegression()
reg.fit(x, y)
r_sq = reg.score(x, y)
p_val = f_regression(x, y)[1][0]
print("The R-squared is {}.".format((r_sq)))
print("The p_value is {}.".format(round(p_val,3)))
```

#### As there is 64% variance in the Cancer Type with respect to concavity error. Therefore, H0 is rejected.

## Initialize features into X variable

```python
X = df.drop(['Cancer_type'], axis=1)
X.head()
```

## Initialize target into y variable

```python
y = df.Cancer_type
y.head()
```

## Heatmap visualization

```python
f,ax = plt.subplots(figsize=(18, 18))
matrix = np.triu(X.corr())
sns.heatmap(X.corr(), annot=True, linewidths=0.5,ax=ax, mask=matrix)
plt.show()
```

```python
plot_data = (df.set_index('Cancer_type').stack().to_frame().reset_index().rename(columns={0:'size', 'level_1':'measurement'}))
plot_data.head()
```

## Split dataset into train and test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

## Shape of train and test dataset

```python
print("X_train shape is: ",X_train.shape)
print("X_test shape is: ",X_test.shape)
```

## Feature scaling of the data

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler(feature_range=(0,1))
X_train = mms.fit_transform(X_train)
X_test = mms.fit_transform(X_test)
```

## to visualize complete values for below command

```python
np.set_printoptions(edgeitems=50)
```

```python
X_train
```

## Importing SVM for modelling

```python
from sklearn.svm import SVC
clf = SVC(kernel="linear")
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

## Importing some metrics components

```python
from sklearn.metrics import accuracy_score, classification_report, mean_squared_error
```

## Computing accuracy score

```python
accuracy_score(y_test, y_pred)
```

## Computing rmse

```python
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
rmse
```

## Classification report

```python
classification_report(y_test, y_pred)
```

## Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```
